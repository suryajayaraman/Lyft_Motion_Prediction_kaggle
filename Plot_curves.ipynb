```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plot_trainloss(file_dir, filename, sample_rate, save_plot=False):
    df = pd.read_csv(file_dir + filename)  
    #print(df.shape)
    sampled_df = df.iloc[::sample_rate, :]
    #print(sampled_df.shape)
    plt.figure(figsize=(10,10))
    plt.plot(sampled_df['iteration'].values[1:] * 16, sampled_df['losses_train'].values[1:]);
    plt.title(filename.replace('.csv', ''))
    plt.xlabel('Number of train Samples')
    plt.ylabel('Train loss')
    plt.show()
    
    if save_plot:
        plt.savefig(filename.replace('.csv', '') + '.png')
```

```python
data_path = '/home/wabco/Downloads/sdxc/Git repos/Lyft_Motion_Prediction_kaggle/submissions/LSTM_Baseline/'
file1_name = 'LSTM_baseline_R34_pvt_10_224_nll_multimode_TRAIN_LOSSES_156k-340k.csv'
file2_name = 'LSTM_baseline_R34_pvt_10_224_nll_multimode_340k.csv'
plot_trainloss(data_path, file1_name, 50)
```
